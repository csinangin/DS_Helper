# **Support Vector Machine (SVM)**

SVM, en önemli ve en güçlü Supervised Machine Learning (Gözetimli Öğrenme) tekniklerinden biridir ve çok yaygın kullanılır.

SVM, özellikle sınıflandırma (ama aynı zamanda regresyon) problemleri için kullanılan denetimli bir öğrenme modelidir. 

SVM'in temel amacı, iki sınıf arasındaki en büyük marjı sağlayan bir ayrım **hyperplane** bulmaktır. 

Bu hyperplane, iki sınıfı birbirinden ayırırken, onları en iyi şekilde ayırt edebilen bir çizgidir (2 boyutta). 

Daha yüksek boyutlarda, bu 'çizgi' bir hyperplane olur.

## **Temel Kavramlar**



**Support Vectors (Destek Vektörleri)**:
- Destek vektörleri, iki sınıf arasındaki hyperplane'i tanımlayan veri noktalarıdır. Yani, sınıflandırma marjını en iyi ayırt eden veri noktalarıdır. Bu noktalar, modelin karar sınırını belirlemede kritik rol oynar.

**Margin (Marj)**:
- Marj, iki sınıf arasındaki mesafedir. SVM, bu marjı en büyük yapan hyperplane'i bulmayı amaçlar. Maksimum marj, modelin genelleme yeteneğini artırır, çünkü sınıflar arasındaki mesafenin maksimum olması, modelin yeni verilerle daha iyi performans göstermesini sağlar.

**Hyperplane**:
- Hyperplane, verileri sınıflandırmak için kullanılan en iyi ayrım çizgisidir. Bu çizgi, iki sınıfı en iyi şekilde ayıran ve sınıflar arasındaki mesafeyi (marjı) en büyük yapan düzlemdir. 2 boyutlu bir uzayda bu bir çizgi, 3 boyutlu bir uzayda bir düzlem ve daha yüksek boyutlarda bir hyperplane olur.

**Kernel (Çekirdek Fonksiyonu)**:
- Kernel fonksiyonları, doğrusal olmayan veri setlerini daha yüksek boyutlu bir uzaya dönüştürerek doğrusal olarak ayırmayı mümkün kılar. Bu, karmaşık ve doğrusal olmayan ilişkileri modellemeye olanak tanır. Yaygın olarak kullanılan kernel fonksiyonları arasında doğrusal, RBF (Radial Basis Function), polinom ve sigmoid bulunur.



## **Avantajları**

1. **Etkili Sınıflandırma**:
   - SVM, küçük ve orta büyüklükteki veri setlerinde yüksek performans gösterir. Özellikle temiz ve gürültüsüz veri setlerinde etkilidir.
  
2. **Yüksek Boyutlu Verilerle Çalışma**:
   - Yüksek boyutlu veri setlerinde de çalışabilir. Özellikle özellik sayısının fazla olduğu durumlarda etkilidir.

3. **Overfitting'e Karşı Dayanıklılık**:
   - SVM, düzenlendirme (regularization) parametreleri ile overfitting'e karşı dirençlidir. Bu, modelin genelleme yeteneğini artırır.

## **Dezavantajları**

1. **Büyük Veri Setlerinde Eğitim Süresi**:
   - Büyük veri setlerinde eğitim süresi uzun olabilir. Bu, hesaplama kaynaklarının yoğun kullanımına neden olabilir.

2. **Özellik ve Kernel Seçimi**:
   - Özellik seçimi ve kernel seçimi, modelin performansı üzerinde büyük bir etkiye sahip olabilir. Doğru seçimler yapılmazsa model performansı düşebilir.

## **Multiclass Classification (Çok Sınıflı Sınıflandırma)**

SVM algoritması, iki sınıflı sınıflandırma için tasarlanmıştır. Ancak, çok sınıflı sınıflandırma problemleri için de kullanılabilir. SVM, çok sınıflı sınıflandırma problemlerinde birden fazla iki sınıflı sınıflandırıcı kullanarak çalışır. Bu yaklaşım, "one-vs-all (rest)" (OvA) veya "one-vs-one" (OvO) olarak bilinir.

**OvA (One-vs-All)**:
- Her sınıf, geri kalan tüm sınıflardan ayrı olarak ele alınır ve sınıflandırıcı, her sınıfı diğerlerinden ayırt etmeye çalışır.
- Örneğin, 3 sınıflı bir problem için, önce sınıf 1'i diğerlerinden ayırmaya çalışan bir sınıflandırıcı oluşturulur, ardından sınıf 2'yi diğerlerinden ayırmaya çalışan bir sınıflandırıcı ve son olarak sınıf 3'ü diğerlerinden ayırmaya çalışan bir sınıflandırıcı oluşturulur.

**OvO (One-vs-One)**:
- Her iki sınıf arasında bir sınıflandırıcı oluşturulur ve tüm sınıflandırıcıların çıktıları bir araya getirilerek en sık tekrarlanan sınıf belirlenir.
- Örneğin, 3 sınıflı bir problem için, sınıf 1 ve sınıf 2 arasında bir sınıflandırıcı, sınıf 1 ve sınıf 3 arasında bir sınıflandırıcı ve sınıf 2 ve sınıf 3 arasında bir sınıflandırıcı oluşturulur. Sonuç olarak, her bir çift sınıf arasındaki sınıflandırıcının çıktıları birleştirilir ve en sık tekrarlanan sınıf belirlenir.

## **Sonuç**

SVM, güçlü ve esnek bir sınıflandırma algoritmasıdır. 

Küçük ve orta büyüklükteki veri setlerinde yüksek performans gösterir ve overfitting'e karşı dayanıklıdır. 

Kernel fonksiyonları sayesinde doğrusal olmayan verileri de başarıyla sınıflandırabilir. 

Çok sınıflı sınıflandırma problemlerinde OvA ve OvO yaklaşımlarıyla geniş bir kullanım alanına sahiptir.

# **SVM Hiperparametreleri**

**C**:
- **Tanım**: Regülarizasyon parametresidir.
- **İşlevi**: Sınıflandırma hatasını minimize ederken overfitting riskini kontrol eder.
- **Özellikleri**:
  - Büyük bir C değeri: Daha düşük sınıflandırma hatası, ancak overfitting riski artar.
  - Küçük bir C değeri: Daha yüksek genelleştirme performansı, overfitting riski azalır.
- **Kullanım Durumu**: Model overfit olmuşsa C’yi azaltmak gerekir.

**Gamma**:
- **Tanım**: Kernel fonksiyonunun parametresidir.
- **İşlevi**: Verinin nonlineer olarak ayrılabilmesini sağlar.
- **Özellikleri**:
  - Düşük gamma: Daha düzgün bir karar sınırı çizer, iki noktanın birbirine yakın olmasını tanımlar.
  - Yüksek gamma: Daha keskin bir karar sınırı çizer, sadece çok yakın noktalar birbirini etkiler.
- **Ölçeklendirme**:
  - **'scale'**: Verilerin ölçeklendirilmesi ile farklı ölçekteki her bir özellik gamma için aynı ağırlığa sahip olur.
  - **'auto'**: 1 / (n_features) olarak hesaplanır.
- **Kullanım Durumu**: Model overfit olmuşsa gamma değerini düşürmek, model underfit olmuşsa gamma değerini yükseltmek gerekir.

**Kernel**:
- **Tanım**: Verilerin daha yüksek boyutlu uzaylara dönüştürülmesinde kullanılan fonksiyondur.
- **Çeşitler**:
  - **Linear**: Veri doğrusal olarak ayrılabiliyorsa kullanılır.
  - **RBF (Radial Basis Function)**: Doğrusal olmayan veri setleri için en çok kullanılan kernel tipidir.
  - **Polynomial**: RBF'den farklı bir yaklaşım sunar, ancak derecesi arttıkça karmaşıklık ve eğitim süresi artar.
  - **Sigmoid**: Özel durumlarda kullanılır, genellikle diğer çekirdeklere kıyasla daha az tercih edilir.
  - **Precomputed**: Önceden hesaplanmış bir çekirdek matrisi kullanır.

**Degree**:
- **Tanım**: Polynomial kernel için derece parametresidir.
- **Özellikleri**:
  - Varsayılan değeri 3’tür.
  - Diğer çekirdeklerde genellikle varsayılan değeri 1’dir.
- **Kullanım Durumu**: Poly çekirdek aktifken genellikle 3 olarak kullanılır.

**Özet**:
- **C** ve **gamma** parametreleri, SVM'nin performansını ve genelleme kabiliyetini doğrudan etkiler.
- **Kernel** seçimi, verinin doğasına ve problemine bağlı olarak yapılmalıdır.
- **Degree**, yalnızca Polynomial kernel için geçerli olup, modelin karmaşıklığını belirler.

Bu parametreler SVM modellerinin performansını optimize etmek için dikkatle seçilmelidir. Overfitting ve underfitting gibi problemleri önlemek için doğru C ve gamma değerlerinin belirlenmesi kritik öneme sahiptir.

## **C Parametresi**

**SVM'de C parametresi, modelin performansını ve davranışını belirlemede kritik bir rol oynar. Esas olarak düşük eğitim hatası ile düşük test hatası arasında denge kurar, bu da modelin genelleme yeteneği için anahtardır.**

**C Parametresi Nedir?**

- **Regularizasyon Parametresi**: C parametresi bir regularizasyon parametresidir. Karar sınırının basitliği ile eğitim örneklerinin yanlış sınıflandırılması arasındaki dengeyi kontrol eder.
- **Aralık**: C, 0 ile sonsuz arasında herhangi bir pozitif değeri alabilir.
- **Ters Orantı**: Regularizasyonun gücü, C'ye ters orantılıdır. Bu nedenle, küçük bir C değeri daha güçlü regularizasyon anlamına gelir.

**C Modeli Nasıl Etkiler?**

1. **Küçük C Değeri**:
    - **Yüksek Regularizasyon**: Küçük bir C değeri, yüksek düzeyde regularizasyon uygular. Bu, modelin daha basit bir karar sınırını tercih etmesine neden olur, hatta bazı eğitim noktalarının yanlış sınıflandırılmasına izin verir.
    - **Genelleme**: Bu genellikle yeni, görülmemiş veriler üzerinde daha iyi genelleme ile sonuçlanır, aşırı uydurmayı (overfitting) azaltır.
    - **Marj**: Marj daha büyük olur ve bireysel veri noktalarına daha az duyarlı hale gelir.

2. **Büyük C Değeri**:
    - **Düşük Regularizasyon**: Büyük bir C değeri, daha az regularizasyon anlamına gelir. Model, tüm eğitim örneklerini doğru sınıflandırmaya çalışır, bu da daha karmaşık bir karar sınırıyla sonuçlanır.
    - **Aşırı Uydurma**: Bu, modelin eğitim verileri üzerinde çok iyi performans gösterip yeni veriler üzerinde kötü performans göstermesine yol açabilir.
    - **Marj**: Marj daha küçük olur ve bireysel veri noktalarına daha duyarlı hale gelir, bu da daha karmaşık bir hiper düzlemle sonuçlanır.

## **Kernel (Çekirdek Fonksiyonu)**

Kernel Fonksiyonu, girdi olarak alınan verileri işlenmesi gereken formata dönüştürmek için kullanılan bir yöntemdir. Support Vector Machine'de (SVM) kullanılan matematiksel fonksiyonlar seti, verileri manipüle etme penceresi sağladığı için "Kernel" adı kullanılır. Kernel Fonksiyonu, düşük boyutlu bir uzayda doğrusal olarak ayrılamayan verileri, daha yüksek boyutlu bir uzaya dönüştürerek doğrusal olarak ayırmayı mümkün kılar. Temelde, iki nokta arasında standart özellik boyutunda iç çarpımı döndürür.

**Kernel Seçimi**

**rbf - Radial Basis Function (Radyal Temel Fonksiyonu)**:
- **Tanım**: RBF kerneli, SVM'lerde en yaygın kullanılan çekirdek fonksiyonudur. Verileri sonsuz boyutlu bir uzaya dönüştürerek, karmaşık ve doğrusal olmayan ilişkileri modellemeye olanak tanır.
- **Parametreler**:
  - **C**: Tüm SVM çekirdeklerinde olduğu gibi, C parametresi eğitim örneklerinin yanlış sınıflandırılması ile karar yüzeyinin basitliği arasında bir denge sağlar. 
    - **Düşük C**: Karar yüzeyini düzgün yapar, overfitting riskini azaltır.
    - **Yüksek C**: Eğitim örneklerinin doğru sınıflandırılmasını hedefler, ancak overfitting riski artar.
  - **Gamma**: Bir eğitim örneğinin ne kadar etkili olduğunu tanımlar.
    - **Düşük Gamma**: Daha geniş bir etki alanı, daha düzgün bir karar sınırı.
    - **Yüksek Gamma**: Daha dar bir etki alanı, daha keskin ve dairesel karar sınırları.
- **Kullanım Durumu**: RBF kerneli, genellikle doğrusal olmayan ilişkiler içeren ve yüksek boyutlu verilerde kullanılır. Modelin karmaşıklığını ve eğitim süresini artırabilir, ancak genellikle en iyi sonuçları verir.

**Linear (Doğrusal Kernel)**:
- **Tanım**: Doğrusal kernel, verilerin doğrusal olarak ayrılabildiği durumlarda kullanılır.
- **Parametreler**: C parametresi doğrusal kernelde de önemlidir.
- **Kullanım Durumu**: Yüksek boyutlu verilerde veya verilerin doğrusal olarak ayrılabildiği durumlarda tercih edilir. Daha az hesaplama gerektirir ve daha hızlı çalışır.

**Polynomial (Polinom Kernel)**:
- **Tanım**: Polinom kernel, verileri polinomsal olarak dönüştürür.
- **Parametreler**:
  - **Degree**: Polinomun derecesini belirler.
  - **C ve Coef0**: Karar yüzeyinin esnekliğini artırır.
- **Kullanım Durumu**: Orta derecede doğrusal olmayan ilişkilerde kullanılır. Derecesi arttıkça modelin karmaşıklığı ve eğitim süresi artar.

**Sigmoid (Logistic Regression Kernel)**:
- **Tanım**: Sigmoid kernel, lojistik regresyon modelinde kullanılan bir fonksiyondur.
- **Parametreler**:
  - **C ve Coef0**: Fonksiyonun eğimini ve esnekliğini belirler.
- **Kullanım Durumu**: Özel durumlarda kullanılır, genellikle diğer kernel tiplerine göre daha az tercih edilir.

**Precomputed (Önceden Hesaplanmış Kernel)**:
- **Tanım**: Kernel matrisi önceden hesaplanmış veriler kullanılarak eğitilir.
- **Kullanım Durumu**: Özellikle büyük veri setlerinde ve özel kernel fonksiyonları kullanıldığında tercih edilir.

**Özet**

Kernel fonksiyonları, SVM'lerin esnekliğini ve genelleme kabiliyetini artırarak farklı veri setlerine uyum sağlamasını mümkün kılar. Kernel seçimi, verinin doğasına ve problemine bağlı olarak dikkatle yapılmalıdır. Doğru kernel ve hiperparametrelerin seçimi, modelin performansını ve genelleme yeteneğini önemli ölçüde etkiler.

## **Gamma**

SVM'deki iki temel hiperparametre olan Gamma ve C, modelin performansını önemli ölçüde etkiler. Bu parametrelerin doğru ayarlanması, modelin genelleme yeteneğini ve doğruluğunu artırabilir. İşte bu parametrelerin nasıl etkileştiğine dair detaylı bir açıklama:

**Gamma:**

- **Marjinlerin Genişliği**: Gamma, SVM'deki marjinlerin genişliğini belirler. Yüksek bir Gamma değeri, sınıfların etrafındaki alanı daraltır, düşük bir Gamma değeri ise genişletir.
- **Sınıflandırma**: Amaç, mümkün olduğunca çok sayıda gözlemi bu alana dahil ederek doğru sınıflandırmayı sağlamaktır. Genellikle, bu alan biraz geniş çizilir ki daha fazla veri noktası dahil edilebilsin.
- **Etkisi**: Gamma değeri arttıkça, model daha keskin ve karmaşık karar sınırları çizer, bu da modelin overfitting (aşırı uyum) yapma riskini artırabilir.

**C:**

- **Regularizasyon**: C, SVM'deki regularizasyon parametresidir. C'nin düşük değerleri, modelin daha basit ve genelleyici olmasını sağlar, yüksek değerleri ise modelin eğitim verisine daha sıkı uyum sağlamasına neden olur.
- **Hata Cezalandırma**: C, modelin eğitim verisindeki hataları ne kadar cezalandıracağını belirler. Düşük C, daha fazla hatayı tolere ederken, yüksek C daha az hatayı tolere eder.

**C ve Gamma'nın Etkileşimi:**

- **Regularizasyon ve Genel Etki**: C ve Gamma değerleri küçüldükçe, regularizasyon katsayısı artar, bu da modelin daha genelleyici olmasını sağlar. Yüksek Gamma değerleri ise modelin daha karmaşık ve keskin karar sınırları çizmesine neden olabilir.
- **Etkileşim**: Gamma değeri arttıkça, C'nin etkisi azalır. Yüksek bir Gamma, sınırlı bir veri kümesi üzerinde daha küçük bir alanı tanımlar, bu da C'nin regularizasyon etkisini azaltır.
- **Grid Search**: C ve Gamma parametrelerinin uygun değerlerini bulmak için genellikle **grid search** (ızgara arama) yöntemi kullanılır. Bu yöntem, farklı C ve Gamma kombinasyonlarını deneyerek en iyi performansı veren parametre çiftini belirler.

**C ve Gamma ile Regularizasyon:**

1. **Düşük C ve Düşük Gamma**:
   - **Regularizasyon Katsayısı**: Yüksek
   - **Model**: Daha basit ve genelleyici
   - **Overfitting Riski**: Düşük

2. **Yüksek C ve Yüksek Gamma**:
   - **Regularizasyon Katsayısı**: Düşük
   - **Model**: Daha karmaşık ve veri setine sıkı uyumlu
   - **Overfitting Riski**: Yüksek

3. **Düşük C ve Yüksek Gamma**:
   - **Model**: Daha karmaşık karar sınırları, fakat genel olarak daha az regularizasyon

4. **Yüksek C ve Düşük Gamma**:
   - **Model**: Daha basit karar sınırları, fakat eğitim verisine sıkı uyum


**Özet**

- **C ve Gamma**: SVM'de modelin performansını belirleyen iki kritik parametredir.
- **Regularizasyon**: Düşük C ve Gamma değerleri, yüksek regularizasyon ve daha basit model; yüksek C ve Gamma değerleri ise düşük regularizasyon ve daha karmaşık model anlamına gelir.
- **Etkileşim**: Gamma değeri arttıkça, C'nin etkisi azalır. Bu etkileşimi anlamak, modelin doğru şekilde ayarlanmasını sağlar.
- **Grid Search**: Parametrelerin en iyi kombinasyonunu bulmak için kullanılır.

## **Degree (Yalnızca Poly Kerneller için)**

SVM'de **Degree** parametresi, yalnızca **polynomial** (polinom) kernel fonksiyonu kullanıldığında anlamlıdır. Diğer tüm kernel türleri tarafından yok sayılır. Bu parametre, polinomun derecesini belirler ve karar sınırının karmaşıklığını doğrudan etkiler.

**Degree Nedir?**

- **Polynomial Kernel**: Polynomial kernel fonksiyonu, verileri daha yüksek boyutlu bir uzaya projelendirmek için polinom terimleri kullanır. Bu, doğrusal olmayan verileri daha kolay ayrılabilir hale getirir.
- **Derece**: Degree, polynomial kernel'in derecesidir. Yani, polinomun üst derecesini temsil eder. Örneğin, degree = 2 ise, ikinci dereceden bir polinom kullanılır.

**Degree'nin Model Üzerindeki Etkisi**

1. **Düşük Degree (Örneğin, 2 veya 3)**:
    - **Karar Sınırı**: Daha az karmaşık bir karar sınırı oluşturur.
    - **Model**: Daha basit ve genelleyici olur, aşırı uydurma (overfitting) riski azalır.
    - **Hesaplama**: Daha az hesaplama gerektirir ve daha hızlı çalışır.

2. **Yüksek Degree (Örneğin, 5 veya Daha Fazlası)**:
    - **Karar Sınırı**: Daha karmaşık ve eğimli bir karar sınırı oluşturur.
    - **Model**: Eğitim verisine daha sıkı uyum sağlar, bu da aşırı uydurma riskini artırabilir.
    - **Hesaplama**: Daha fazla hesaplama gerektirir ve eğitim süresi uzar.

**Özet**

- **Degree**: Polynomial kernel fonksiyonunun derecesidir. Yalnızca polynomial kernel kullanıldığında anlamlıdır.
- **Düşük Degree**: Daha basit ve genelleyici model, daha az hesaplama.
- **Yüksek Degree**: Daha karmaşık ve eğilimli model, aşırı uydurma riski, daha fazla hesaplama.
- **Grid Search**: Degree ve diğer parametrelerin en iyi kombinasyonunu bulmak için kullanılır.

Degree parametresinin doğru ayarlanması, polynomial kernel kullanan SVM modellerinde önemli bir performans farkı yaratabilir. Bu parametre, karar sınırının karmaşıklığını ve modelin genelleme yeteneğini doğrudan etkiler.
